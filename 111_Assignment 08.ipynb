```python
# task 1

class RealNumber: 
    
    def __init__(self, r=0): 
        self.__realValue = r 
    def getRealValue(self): 
        return self.__realValue 
    def setRealValue(self, r): 
        self.__realValue = r 
    def __str__(self): 
        return 'RealPart: '+str(self.getRealValue())
    
class ComplexNumber(RealNumber):
    def __init__(self,r=1,i=1):
        super().__init__(r)
        self.i=i
    def __str__(self): 
        return 'RealPart: '+str(self.getRealValue())+'.0'+'\n'+'ImaginaryPart: '+str(self.i)+'.0'
    
    
cn1 = ComplexNumber() 
print(cn1) 
print('---------') 
cn2 = ComplexNumber(5,7) 
print(cn2)    
```

```python
# task 2

class RealNumber: 
    def __init__(self, number=0): 
        self.number = number 
    def __add__(self, anotherRealNumber): 
        return self.number + anotherRealNumber.number 
    def __sub__(self, anotherRealNumber): 
        return self.number - anotherRealNumber.number 
    def __str__(self): 
        return str(self.number)
    
class ComplexNumber(RealNumber):
    def __init__(self,number,anotherRealNumber):
        super().__init__(number)
        self.anotherRealNumber=anotherRealNumber
        
    def __add__(self, anotherRealNumber): 
        a=int(str(anotherRealNumber.number))
        b=self.number+a
        c=int(str(anotherRealNumber.anotherRealNumber))
        d=self.anotherRealNumber+c
        return str(b)+' + '+str(d)+'i'
    
    def __sub__(self, anotherRealNumber): 
        a=int(str(anotherRealNumber.number))
        b=self.number-a
        c=int(str(anotherRealNumber.anotherRealNumber))
        d=self.anotherRealNumber-c
        return str(b)+' '+str(d)+'i'
    
    def __str__(self):
        return str(self.number)+' + '+str(self.anotherRealNumber)+'i'
    
r1 = RealNumber(3) 
r2 = RealNumber(5) 
print(r1+r2) 
cn1 = ComplexNumber(2, 1) 
print(cn1) 
cn2 = ComplexNumber(r1, 5) 
print(cn2) 
cn3 = cn1 + cn2 
print(cn3) 
cn4 = cn1 - cn2 
print(cn4)
```

```python
# task 3

class Account: 
    def __init__(self, balance): 
        self._balance = balance 
        
    def getBalance(self): 
        return self._balance
    
class CheckingAccount(Account):
    numberOfAccount=0
    def __init__(self,balance=0):
        super().__init__(float(balance))
        CheckingAccount.numberOfAccount+=1
        
    def __str__(self):
        return 'Account Balance: '+str(super().getBalance())
        
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount) 
print(CheckingAccount()) 
print(CheckingAccount(100.00)) 
print(CheckingAccount(200.00))
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)    
```

```python
# task 5

class Exam: 
    def __init__(self,marks): 
        self.marks = marks 
        self.time = 60 
        
    def examSyllabus(self): 
        return "Maths , English" 
    def examParts(self): 
        return "Part 1 - Maths\nPart 2 - English\n"
    
class ScienceExam(Exam):
    def __init__(self,marks,time,*syllabus):
        super().__init__(marks)
        self.time=time
        if len(syllabus)==2:
            self.syllabus1=syllabus
            self.syllabus2=syllabus
            self.p=4
            
        elif len(syllabus)==3:
            self.syllabus1=syllabus
            self.syllabus2=syllabus
            self.syllabus3=syllabus
            self.p=5
            
    def __str__(self):
        if self.p==4:
            return 'Marks: '+str(self.marks)+' '+'Time: '+str(self.time)+' minutes'+' Number of Parts: '+str(self.p)
        elif self.p==5:
            return 'Marks: '+str(self.marks)+' '+'Time: '+str(self.time)+' minutes'+' Number of Parts: '+str(self.p)
        
    def examSyllabus(self):
        if self.p==4:
            return super().examSyllabus()+','+' Physics'+','+' HigherMaths'
        elif self.p==5:
            return super().examSyllabus()+','+' Physics'+','+' HigherMaths'+','+' Drawing'
        
    def examParts(self):
        if self.p==4:
            return super().examParts()+'Part 3 - Physics\nPart 4 - HigherMaths'
        elif self.p==5:
            return super().examParts()+'Part 3 - Physics\nPart 4 - HigherMaths\nPart 5 - Drawing'
    
engineering = ScienceExam(100,90,"Physics","HigherMaths") 
print(engineering) 
print('----------------------------------') 
print(engineering.examSyllabus()) 
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing") 
print(architecture) 
print('----------------------------------') 
print(architecture.examSyllabus()) 
print(architecture.examParts())
```

```python
# task 6

class Shape3D: 
    pi = 3.14159 
    def __init__(self, name = 'Default', radius = 0): 
        self._area = 0 
        self._name = name 
        self._height = 'No need' 
        self._radius = radius 
        
    def calc_surface_area(self): 
        return 2 * Shape3D.pi * self._radius
    
    def __str__(self): 
        return "Radius: "+str(self._radius)
    
class Sphere(Shape3D):
    def __init__(self,name,radius):
        super().__init__(name,radius)
        self._height = 'No need'
        return 'Shape name: '+str(self.name)+','+'Area Formula: '+'4 * pi * r * r'
    
    def calc_surface_area(self): 
        self.area=2 * Shape3D.pi * self._radius
        return 'Radius: '+str(self.radius)+','+'Height: '+str(self.height)+'\n'+'Area: '+str(self.area)
    
class Cylinder(Shape3D):
    
    def __init__(self,name,radius,height):
        super().__init__(name,radius)
        self.height=height
        return 'Shape name: '+str(self.name)+','+'Area Formula: '+'4 * pi * r * (r+h)'
    
    def calc_surface_area(self): 
        self.area=2 * Shape3D.pi * self._radius
        return 'Radius: '+str(self.radius)+','+'Height: '+str(self.height)+'\n'+'Area: '+str(self.area)
    
sph = Sphere('Sphere', 5) 
print('----------------------------------') 
sph.calc_surface_area() 
print(sph) 
print('==================================') 
cyl = Cylinder('Cylinder', 5, 10) 
print('----------------------------------') 
cyl.calc_surface_area() 
print(cyl)    
```
